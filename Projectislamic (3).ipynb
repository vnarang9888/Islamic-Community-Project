```python

import pandas as pd
df= pd.read_csv('mergedfoodandclients.csv')
```

```python
print(df_selected.dtypes)
for column in df_selected.columns:
  print(f"Column '{column}' has data type: {df_selected[column].dtype}")
```

```python
import pandas as pd
# Convert 'pickup_date' and 'datetime_from' columns to datetime objects
df_selected['pickup_date'] = pd.to_datetime(df_selected['pickup_date'], errors='coerce')
df_selected['datetime_from'] = pd.to_datetime(df_selected['datetime_from'], errors='coerce')

df_selected['collect_scheduled_date'] = pd.to_datetime(df_selected['collect_scheduled_date'], errors='coerce')
print(df_selected.dtypes)
```

```python
# Rename the 'collect_scheduled_date' column to 'schedule_date'
df_selected = df_selected.rename(columns={'collect_scheduled_date': 'schedule_date'})

# Count the number of events for each date
pickup_series = df_selected.groupby(df_selected['pickup_date'].dt.date).size().rename('actual_pickup')
schedule_series = df_selected.groupby(df_selected['schedule_date'].dt.date).size().rename('scheduled_pickup')

print(pickup_series.head())
print(schedule_series.head())
```

Defining the time range for prediction

```python
start_date = pd.to_datetime('2023-11-01')
end_date = pd.to_datetime('2024-08-26')
prediction_dates=pd.date_range(start=start_date, end=end_date, freq='D')
```

Reindex the time series to the full date range and fill missing values with 0

```python
pickup_series = pickup_series.reindex(prediction_dates, fill_value=0)
schedule_series = schedule_series.reindex(prediction_dates, fill_value=0)
```

```python

# Create the merged time series DataFrame
merged_time_series = pd.DataFrame({
    'date': prediction_dates,
    'actual_pickup': pickup_series.values,
    'scheduled_pickup': schedule_series.values
})

print(merged_time_series.head())
```

```python

import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import acf

# Assuming 'df' is your merged_time_series DataFrame
# Calculate the cross-correlation function (ACF)
lags = range(len(df))  # Consider all possible lags
acf_values = acf(df['actual_pickup'], nlags=len(df) - 1)

# Find the lag with the highest absolute correlation (excluding lag 0)
highest_correlation = 0
highest_lag = 0
for lag, correlation in enumerate(acf_values[1:]):  # Skip lag 0
    if abs(correlation) > abs(highest_correlation):
        highest_correlation = correlation
        highest_lag = lag + 1  # Adjust lag index


# Plot the line graph
plt.figure(figsize=(12, 6))
plt.plot(lags[1:], acf_values[1:]) #excluding lag 0
plt.xlabel('Lag')
plt.ylabel('Correlation')
plt.title('Autocorrelation Function (ACF) of Actual Pickup')
plt.axvline(x=highest_lag, color='r', linestyle='--', label=f'Highest correlation at lag {highest_lag} (Correlation = {highest_correlation:.2f})')
plt.legend()
plt.grid(True)
plt.show()

print(f"Lag with highest correlation: {highest_lag}")
print(f"Correlation value: {highest_correlation}")
```

```python

import matplotlib.pyplot as plt
import statsmodels.api as sm
from sklearn.metrics import r2_score
from scipy import stats
from scipy.special import inv_boxcox

# Apply Box-Cox transformation to the actual pickup series
# Add 1 to avoid negative or zero values in the pickup count
df['actual_pickup_boxcox'], lam = stats.boxcox(df['actual_pickup'] + 1)
print(f"Box-Cox Lambda: {lam}")
```

```python

# Plot the transformed series to inspect variance stabilization
plt.figure(figsize=(10, 6))
plt.plot(df['date'], df['actual_pickup_boxcox'], label='Box-Cox Transformed Actual Pickups')
plt.title('Actual Pickup Time Series After Box-Cox Transformation')
plt.xlabel('Date')
plt.ylabel('Transformed Pickup Count')
plt.grid(True)
plt.legend()
plt.show()
```

```python

# Calculate the rolling standard deviation (window=30 days)
df['rolling_std_original'] = df['actual_pickup'].rolling(window=30).std()
df['rolling_std_boxcox'] = df['actual_pickup_boxcox'].rolling(window=30).std()

# Step 2: Plot the rolling standard deviation before and after Box-Cox
plt.figure(figsize=(10, 6))
plt.plot(df['date'], df['rolling_std_original'], label='Rolling Std (Original)')
plt.plot(df['date'], df['rolling_std_boxcox'], label='Rolling Std (Box-Cox)', color='red')
plt.title('Rolling Standard Deviation (30-day window)')
plt.xlabel('Date')
plt.ylabel('Rolling Standard Deviation')
plt.legend()
plt.grid(True)
plt.show()
```

```python

df['scheduled_pickup_lag_7'] = df['scheduled_pickup'].shift(-7)
df['scheduled_pickup_lag_14'] = df['scheduled_pickup'].shift(-14)
# Forward fill missing values in exogenous variables
df[['scheduled_pickup', 'scheduled_pickup_lag_7', 'scheduled_pickup_lag_14']] = df[['scheduled_pickup', 'scheduled_pickup_lag_7', 'scheduled_pickup_lag_14']].ffill()

# Step 3: Split the data into training and test sets (80% train, 20% test)
split_idx = int(len(df) * 0.8)  # 80% of the data for training
train_df = df.iloc[:split_idx]
test_df = df.iloc[split_idx:]
```

```python

exog_train = train_df[['scheduled_pickup', 'scheduled_pickup_lag_7', 'scheduled_pickup_lag_14']]

# Test data
exog_test = test_df[['scheduled_pickup', 'scheduled_pickup_lag_7', 'scheduled_pickup_lag_14']]
```

```python

# Step 4: Fit SARIMA Model on training data using Box-Cox transformed data
sarima_model_boxcox = sm.tsa.SARIMAX(train_df['actual_pickup_boxcox'],
                                    exog=exog_train,  # Use schedule_count as an external regressor
                                    order=(4, 1, 4),  # ARIMA(p,d, q)
                                    seasonal_order=(1, 1, 1, 7))  # SARIMA with seasonal component (weekly seasonality)
sarima_fit_boxcox = sarima_model_boxcox.fit(disp=False)
print(sarima_fit_boxcox.summary())
```

```python

# Step 5: Forecasting the test data period using the model
forecast_boxcox = sarima_fit_boxcox.get_forecast(steps=len(test_df), exog=exog_test)
forecast_values_boxcox = forecast_boxcox.predicted_mean
confidence_intervals_boxcox = forecast_boxcox.conf_int()
```

```python

# Step 6: Plotting the forecast (Box-Cox transformed data)
plt.figure(figsize=(12, 6))
plt.plot(train_df['date'], train_df['actual_pickup_boxcox'], label='Box-Cox Transformed Actual Pickups (Train)')
plt.plot(test_df['date'], test_df['actual_pickup_boxcox'], label='Box-Cox Transformed Actual Pickups (Test)')
plt.plot(test_df['date'], forecast_values_boxcox, label='Forecasted Pickups (Box-Cox)')
plt.fill_between(test_df['date'], confidence_intervals_boxcox.iloc[:, 0], confidence_intervals_boxcox.iloc[:, 1], color='gray', alpha=0.3)
plt.title('SARIMA PICKUP FOR BOX COX PICKUPS(TRAIN VS TEST)')
plt.legend()
plt.show()
```

```python
import pandas as pd

# Load the dataset with the appropriate encoding to avoid errors
file_path = '/content/Islamic calendar.csv'
ff = pd.read_csv(file_path,encoding='latin-1')
ff.head()
```

```python
# Rename columns for clarity
ff.columns = ['Date', 'Event', 'Day']
```

```python
# Convert the 'Date' column to a proper datetime format, then format it to a short date (YYYY-MM-DD)
ff['Date'] = pd.to_datetime(ff['Date'], format='%d-%b-%y', errors='coerce')
```

```python
# Step 1: Convert start and end dates to datetime objects and generate a date range
start_date = pd.to_datetime('2023-11-06')  # Start of the prediction period
end_date = pd.to_datetime('2024-08-26')    # End of the prediction period
prediction_dates = pd.date_range(start=start_date, end=end_date, freq='D')  # Generate all dates between start and end with daily frequency

# Step 2: This function will calculate aggregated features (like number of pickups in the past) for each prediction date
def create_aggregated_features(ff, prediction_date):
    windows = [7, 14, 30]  # Define the time windows of interest (past 7 days, 14 days, and 30 days)
    features = {}  # Initialize an empty dictionary to store the features for each window

    # Loop over each time window to calculate features
    for window in windows:
        # Define the start and end of the time window
        time_window_start = prediction_date - pd.Timedelta(days=window)  # Start of the time window (past 'window' days)
        time_window_end = prediction_date - pd.Timedelta(days=1)  # End of the time window (day before the prediction date)

        # Filter the DataFrame to only include rows within the time window
        ff_pickups_past = ff[(ff['Date'] >= time_window_start) & (ff['Date'] < time_window_end)]

        # Count the number of pickups (or events) that occurred within the time window
        features[f'pickups_past_{window}_days'] = ff_pickups_past.shape[0]  # Store the count of pickups in the features dictionary

    return features  # Return the dictionary of features for the given prediction date

# Step 3: Loop over each prediction date and generate the features
aggregated_features_list = []  # Initialize an empty list to store the feature dictionaries for each date

# Step 4: For each prediction date, generate the features and append them to the list
for prediction_date in prediction_dates:
    features = create_aggregated_features(ff, prediction_date)  # Generate features for the current prediction date
    features['prediction_date'] = prediction_date  # Add the prediction date to the features
    aggregated_features_list.append(features)  # Append the feature dictionary to the list

# Step 5: Convert the list of dictionaries into a DataFrame for easy analysis
isc_date = pd.DataFrame(aggregated_features_list)  # Create a DataFrame from the list of features
```

```python
import pandas as pd

# Step 4: Define the time-range for prediction
class PredictionWindow:
    def __init__(self, start, end, frequency='D'):
        self.start_date = pd.to_datetime(start)
        self.end_date = pd.to_datetime(end)
        self.prediction_dates = pd.date_range(start=self.start_date, end=self.end_date, freq=frequency)

# Step 5: Class to handle feature aggregation based on time windows
class FeatureAggregator:
    def __init__(self, dataframe, windows=None):
        self.ff = dataframe
        self.windows = windows if windows else [7, 14, 30]

    def aggregate(self, prediction_date):
        features = {}
        for window in self.windows:
            time_window_start = prediction_date - pd.Timedelta(days=window)
            time_window_end = prediction_date - pd.Timedelta(days=1)

            ff_filtered = self.ff[(self.ff['Date'] >= time_window_start) & (self.ff['Date'] < time_window_end)]
            features[f'pickups_past_{window}_days'] = ff_filtered.shape[0]

        return features

# Example usage for creating features for each prediction date
def create_features_for_predictions(ff, start, end):
    prediction_window = PredictionWindow(start, end)
    aggregator = FeatureAggregator(ff)

    aggregated_features = []
    for date in prediction_window.prediction_dates:
        features = aggregator.aggregate(date)
        features['prediction_date'] = date
        aggregated_features.append(features)

    return pd.DataFrame(aggregated_features)

# Call the function to generate the DataFrame with features
aggregated_features_ff = create_features_for_predictions(ff, '2023-11-06', '2024-08-26')

```
